#### AIRLINE을 기준으로 데이터프레임을 나누고 $\to$ ARR_DELAY에 mean함수를 적용 (AIRLINE $\to$ {ARR_DELAY:mean()})

`-` 방법1

```python
df.groupby(by='AIRLINE').agg({'ARR_DELAY':'mean'}) #mean에서 ''를 없애면 적용이 안된다 => 외부에 없는 함수이기 떼문
```

```python
#df.groupby(by='AIRLINE').agg({'ARR_DELAY':mean}) #mean에서 ''를 없애면 적용이 안된다 => 외부에 없는 함수이기 떼문
```

```python
mean?
```

`-` 방법2 ($\star$$\star$$\star$)

```python
df.groupby(by='AIRLINE').agg({'ARR_DELAY':np.mean})
```

`-` 방법3

```python
df.groupby(by='AIRLINE')['ARR_DELAY'].agg('mean')
```

`-` 방법4 ($\star$)

```python
df.groupby(by='AIRLINE')['ARR_DELAY'].agg(np.mean)
```

`-` 방법5

```python
df.groupby(by='AIRLINE')['ARR_DELAY'].mean() #내장함수를 뒤에 붙여서 이용 가능
```

`-` 방법2와 방법4는 사용자정의 함수를 쓸 수 있다는 중요한 장점이 있음

```python
def f(x):return -np.mean(x)
```

```python
df.groupby(by='AIRLINE').agg({'ARR_DELAY':f})
```

```python
df.groupby(by='AIRLINE').agg({'ARR_DELAY':lambda x:-np.mean(x)})
```

```python
df.groupby(by='AIRLINE')['ARR_DELAY'].agg(lambda x : -np.mean(x))
```

`-` 입력이 여러개인 사용자정의 함수도 사용 가능함

```python
def f(x,y):return np.mean(x)**y
```

```python
df.groupby(by='AIRLINE').agg({'ARR_DELAY': lambda x:f(x,2)}) #lambda 사용 외엔 방법2에 적용 어려움
```

```python
df.groupby(by='AIRLINE')['ARR_DELAY'].agg(f,2) #방법4
```

`-` 목표: DIST를 적당한 구간으로 나누어 카테고리화 하고 그것을 바탕으로 groupby를 수행하자.

```python
df.DIST.describe()
```

`-` 구간을 아래와 같이 설정한다.

```python
bins=[-np.inf,400,700,1200,np.inf]
```

`-` pd.cut()을 이용하며 각 구간의 obs(ervation)를 카테고리화 하자.

```python
cuts=pd.cut(df.DIST,bins=bins)
cuts
```

`-` cuts, AIRLINE $\to$ {DIVERTED: sum} (DIVERTED : 우회하는 경우)

```python
df.groupby([cuts,'AIRLINE']).agg({'DIVERTED':'sum'})
```

`-` 아래와 비교해보자.

```python
df.groupby(['AIRLINE']).agg({'DIVERTED':'sum'})
```

`-`cuts를 이용하여 추가그룹핑을 하면 조금 다른 특징들을 데이터에서 발견할 수 있다.
- AA항공사와 DL항공사는 모두 비슷한 우회횟수를 가지고 있음.
- AA항공사는 700회 이상의 구간에서 우회를 많이하고 DL항공사는 400~700사이에서 우회를 많이 한다. (패턴이 다름)

`-` 구간이름에 label을 붙이는 방법

```python
cuts2=pd.cut(df.DIST,bins=bins,labels=['Q1','Q2','Q3','Q4'])
cuts2
```

```python
df.groupby(by=[cuts2,'AIRLINE']).agg({'DIVERTED':sum})
```

```python
df.groupby(cuts2).agg({'DIVERTED':len})
```

숙제

구간을

bins[-np.inf, 400, 700, 1200, np.inf]

이 아니라

bins=[-np.inf,400,600,800,1000,1200,np.inf]

와 같이 나누고 적당한 각 구간별로 해당하는 관측치의 수를 구하라.
